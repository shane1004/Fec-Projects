```
import pandas as pd
maruti = pd.read_csv(r'/content/Quote-Equity-MARUTI-EQ-01-01-2019-to-31-12-2019.csv')
tata = pd.read_csv(r'/content/Quote-Equity-TATAMOTORS-EQ-01-01-2019-to-31-12-2019.csv')
```

```
tata['Date '] = pd.to_datetime(tata['Date '])
maruti['Date '] = pd.to_datetime(maruti['Date '])
```

```
tata.set_index('Date ',inplace=True)
maruti.set_index('Date ', inplace=True)
```

```
tataprice = tata['close ']
import matplotlib.pyplot as plt
tataprice.plot()
plt.title('Tata')
plt.xlabel('Time')
plt.ylabel('Price')
```

```
tatareturns= (tata['close '].pct_change()[1:])*100
tatareturns.plot()
plt.title('Tata')
plt.xlabel('Time')
plt.ylabel('Returns')
```

```
maruti['close '] = maruti['close '].str.replace(',', '').astype(float)
```

```
marutireturns= (maruti['close '].pct_change()[1:])*100
marutireturns.plot()
plt.title('Maruti')
plt.xlabel('Time')
plt.ylabel('Returns')
```

```
marutireturns.plot(label='maruti')
tatareturns.plot(label='tata')
plt.title('Comparing returns')
plt.xlabel('Time')
plt.ylabel('Returns in %')
plt.legend()
```

```
import numpy as np
import scipy.stats as stats
print ('Statistical Moments of Tata')
print ('Mean of the return:',np.mean(tatareturns))
print ('Variance of return:',np.var(tatareturns))
print ('Skewness of return:',stats.skew(tatareturns))
print ('Kurtosis of return:',stats.kurtosis(tatareturns))
```

```
print ('Statistical Moments of Maruti')
print ('Mean of the return:',np.mean(marutireturns))
print ('Variance of return:',np.var(marutireturns))
print ('Skewness of return:',stats.skew(marutireturns))
print ('Kurtosis of return:',stats.kurtosis(marutireturns))
```

```
r = np.corrcoef(tatareturns,marutireturns)[0,1]
print ('Correlation Coefficient:',r)
```

**Correlation & Causation involved between the stocks of the two companies here:**

The return of the stocks of the two companies show a correlation of around 0.44 this implies that their returns show a linear relationship among one another but this linear relationship does not imply that return percentage of one stock is due to the return of the other. Here the return in both the stocks are caused by some factors which are similar like global markets , supply & demand while some factors which are different like the performance of the company.
Here a positive correlation is observed due to the similar factors affecting both the companies. 

**Why is variance used as a risk measure?**

Variance is the statistical tool that measures the spread of data from the mean of data.
It is often used as a risk measure in financial analysis as it portrays the volatility of data from its mean i.e. it measures how much the data posseses a risk of deviating from it's mean as large deviations imply unstability of the stock.

**Why stock return distribution is not a normal curve?**

Stock return distributions are assumed to be normal distributions but they are not truly normal. Return distributions often have fat tails and narrow peaks than expected from the normal curves because of the fact that extreme events are more likely to occur in real life scenarios than expressed by the normal distribution.
